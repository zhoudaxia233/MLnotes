# 1. Simple Plot

在正式开始之前，先用通俗易懂的语言介绍几个容易混淆的概念。  
1. figure和subplot和axes的区别：  
**figure**: 相当于画架。  
**subplot以及axes**: 相当于画架上面放的画纸。  
那么，subplot和axes有什么区别呢？  
subplot是axes的一个特例，是规规矩矩放在画板中央的画纸，而axes放在画板的哪里都可以。

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()  # 调用subplots方法，创建并返回一个画板还附赠一张画纸
ax.plot(t, s)  # 在画纸上画画

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Sehr einfach')
ax.grid()

fig.savefig("test.png")  # 把画板整理一下，将画板上放的画纸命名然后归档
plt.show()
```

# 2. Multiple subplots in one figure

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0, 50)
x2 = np.linspace(0.0, 2.0, 50)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)  # subplot方法在当前的画板上创建一张画纸；画板被分成2行1列，其中当前的画纸放在1这个位置上。也可以写成plt.subplot(211)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(212)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

# 3. Histogram

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

mu = 100
sigma = 15
x = mu + sigma * np.random.randn(500)

num_bins = 50

fig, ax = plt.subplots()

n, bins, patches = ax.hist(x, num_bins, density=1)  # n: The values of the histogram bins.
                                                    # bins: The edges of the bins.
                                                    # Silent list of individual patches used to create the histogram or list of such list if multiple input datasets.

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu)) ** 2))
ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
